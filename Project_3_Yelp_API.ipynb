```python
# Dependencies
import pandas as pd
import requests
import json
import pprint
import urllib
from urllib.parse import quote
```

```python
# Constants
api_key = 'sp8XYAjUKMfB3Ju6COkBTJUBdCID_zRqvRvyis5_azBv823YKx30j85LTyZl2h6q0p4_CHwLQeFO_n-UXsLRqWFVlPuFq4kXiOWdCald8vg1OxyTtYtas3uMiEgLZnYx'
base_url = 'https://api.yelp.com'
SEARCH_PATH = '/v3/businesses/search'
BUSINESS_PATH = '/v3/businesses/'
```

```python
# Pulling restaurant data for Detroit, Michigan and putting it into a DataFrame

# Parameters
MI_url_params = {
    'term': 'restaurants',
    'location': 'Detriot, Mi',
    'limit': 50}

# Setting up the URL to include our parameters
url = '{0}{1}'.format(base_url, quote(SEARCH_PATH.encode('utf8')))
headers = {'Authorization': 'Bearer %s' % api_key}

# Requesting data to be retreived and converting it into json format
Detroit_Restaurants = requests.get(url, headers=headers, params=MI_url_params).json()

# Setting up the columns for our DataFrame
Detroit_Restaurants_List = []

for venue in Detroit_Restaurants['businesses']:
    name = venue['name']
    address = venue['location']['address1']
    zip_code = venue['location']['zip_code']
    country = venue['location']['country']
    city = venue['location']['city']
    state = venue['location']['state']
    latitude = venue['coordinates']['latitude']
    longitude = venue['coordinates']['longitude']
    rating = venue['rating']
    Detroit_Restaurants_List.append(
        {"name": name, "address": address, 
                                     "zip_code": zip_code, "city": city, "state": state, "country": country, 
                                     "latitude": latitude, "longitude": longitude, 
                                     "rating": rating})

# Creates our DataFrame
Detroit_df = pd.DataFrame(Detroit_Restaurants_List)
Detroit_df.head(20)
```

```python
# Export our DataFrame as a CSV, without the Pandas index, but with the header
Detroit_df.to_csv("c:/Users/jclaw/Desktop/Detroit_df.csv", index=False, header=True)
```
